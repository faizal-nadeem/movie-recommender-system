```python
# in the above output of genres is not understandable and weird so we do it understandable.
# in this form - ['Action','Adventure','Fantasy','Science Fiction']
```

```python
def convert(obj):
    L = []
    for i in obj:
        L.append(i['name'])
    return L
```

```python
# convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
#if i run the above line of code directly this throw an error 
#TypeError: string indices must be integers, not 'str'

# So we convert this, here is a module named 'ast'. with the use of this module we can easily convert it.
```

```python
import ast
```

```python
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

**now we create a method for conversion.**

```python
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(3)
```

```python
# Now do same thing for keywords columns
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(3)
```

```python
# Ok now we do for cast, but in the cast column we want only three cast names so here below is the methode to extract three names from the cast column.
```

```python
def convert3(obj):
    l = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            l.append(i['name'])
            count+=1
        else:
            break
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(3)
```

```python
movies['crew'][0]
```

```python
# In the above output od crew you see the many crew members in the column but we want only director so here is a method to extract director name.
```

```python
def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(3)
```

```python
movies['overview'][0]
```

```python
# In the above output line you see the overview is a string so we convert it to a list.
```

```python
movies['overview'] = movies["overview"].apply(lambda x:x.split())
```

```python
movies.head(3)
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(4)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(5)
```

```python
dataset = movies[['movie_id','title','tags']]
```

```python
dataset.head(10)
```

```python
# You see in tags column data values are in list form convert it to string form.
```

```python
dataset['tags'] = dataset['tags'].apply (lambda x:" ".join(x))
```

```python
dataset.head(5)
```

```python
dataset['tags'][0]
```

```python
dataset['tags'] = dataset['tags'].apply(lambda x:x.lower())
```

```python
dataset.head(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 6000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(dataset['tags']).toarray()
```

```python
vectors[0]
```

```python
for word in cv.get_feature_names_out():
    print(word)
```

```python
import nltk      # it is the famous module to do this work of scenarios.
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
dataset['tags'] = dataset['tags'].apply(stem)
```

```python
print(ps.stem('Loved'), ps.stem('Lover'), ps.stem('Loving'))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

# to make a method name recommended, i give a movie to the method and it gives me 5 similar movies like i'm giving it to the method.

```python
def recommended(movie):
    movie_index = dataset[dataset['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key = lambda x:x[1])[1:6]

    for i in movies_list:
        print(dataset.iloc[i[0]].title)
    return
```

```python
recommended('Avatar')
```
